```python
import re
```

```python
new_data=re.sub('[^a-zA-Z]',' ',data)
```

```python
new_data
```

```python
All_words=nltk.word_tokenize(new_data)
```

```python
All_words
```

```python
cleaned_d=[]
new_word_list=[]
for w in All_words:
  w=w.lower()
  # new_word_list=[]
  if w not in sw:
    lemmed=lem.lemmatize(w)
    new_word_list.append(lemmed)
v=' '.join(new_word_list)
cleaned_d.append(v)
cleaned_d
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vector=CountVectorizer()
input=vector.fit_transform(cleaned_d).toarray()
input
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vector=TfidfVectorizer()
input=vector.fit_transform(cleaned_d).toarray()
input
```
